## Extraction du texte depuis le PDF

```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    """Extrait tout le texte d'un fichier PDF page par page."""
    text = ""
    with open(pdf_path, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            text += page.extract_text() + "\n"
    return text

# --- Charger le PDF arabe ---
PDF_PATH = "nlp_document_text_arabe.pdf"  # Adapter le chemin si besoin
raw_text = extract_text_from_pdf(PDF_PATH)

print("=== TEXTE BRUT EXTRAIT DU PDF ===")
print(raw_text[:800])  # Afficher les 800 premiers caractères
```

## Version 1 : Pipeline NLP Arabe

### 1. Tokenisation (التقطيع)
Le texte arabe est segmenté en unités atomiques (tokens). Contrairement au français, l'arabe est une langue **sémitique sans espacement systématique** des clitiques. On normalise d'abord en supprimant les diacritiques (التشكيل) et les caractères spéciaux, puis on découpe par les espaces.

### 2. Stopwords (التنقية)
NLTK fournit une liste de **stopwords arabes** (مثل : في، من، على، إلى، و، أن). Ces mots grammaticaux sont statistiquement trop fréquents pour discriminer un texte. On les retire pour ne garder que les **mots porteurs de sens**.

### 3. Racinisation — Stemming (الجذر)
L'arabe est une langue à **racines trilitères** : presque tous les mots dérivent d'une racine de 3 consonnes. L'`ISRIStemmer` (NLTK) ramène chaque mot à sa racine morphologique. Par exemple, "الحاسوب" → "حاسوب" et "معالجة" → "عالج".

> **Pourquoi ?** Si un autre texte utilise "الحواسيب" (pluriel) ou "حاسوبية" (adjectif), le modèle reconnaîtra la même racine "حسب", représentant le même concept.

### 4. TF-IDF (التمثيل الرقمي)
Le tableau final montre le **poids mathématique** de chaque mot.
- **TF** : Plus "معالجة" apparaît dans votre texte, plus son score monte.
- **IDF** : Si "اللغات" apparaît dans tous les documents du corpus, son score baisse — il perd son caractère discriminant.

**Résultat final :** Votre phrase arabe est transformée d'une structure linguistique humaine en un **vecteur de probabilités mathématiques**, prêt pour un classifieur (Naive Bayes, SVM, etc.).

```python
import re
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem.isri import ISRIStemmer
from sklearn.feature_extraction.text import TfidfVectorizer

# Outils arabes
arabic_stopwords = set(stopwords.words('arabic'))
stemmer_ar = ISRIStemmer()

# STEP 1 : Normalisation + Tokenisation
def normalize_arabic(text):
    text = re.sub(r'[\u0610-\u061A\u064B-\u065F]', '', text)  # Supprimer diacritiques
    text = re.sub(r'[إأآا]', 'ا', text)   # Normaliser les alifs
    text = re.sub(r'ى', 'ي', text)         # Normaliser ya'
    text = re.sub(r'ة', 'ه', text)         # Normaliser ta' marbuta
    text = re.sub(r'[^\u0600-\u06FF\s]', '', text)  # Garder seulement l'arabe
    return text.strip()

texte_norm = normalize_arabic(phrase_test_ar)
tokens_v1 = texte_norm.split()
print(f"1. Tokenisation :\n{tokens_v1}\n")

# STEP 2 : Retrait des Stopwords
tokens_filtres_v1 = [t for t in tokens_v1 if t not in arabic_stopwords]
supprimes_v1 = [t for t in tokens_v1 if t in arabic_stopwords]
print(f"2. Après Stopwords :\n{tokens_filtres_v1}")
print(f"   (Mots supprimés : {supprimes_v1})\n")

# STEP 3 : Stemming (Racinisation)
racines_v1 = [stemmer_ar.stem(t) for t in tokens_filtres_v1]
print(f"3. Racines (ISRIStemmer) :\n{racines_v1}\n")

# STEP 4 : TF-IDF
corpus_v1 = [
    " ".join(racines_v1),
    "طالب جامعه حاسوب تطوان",
    "لغه طبيعيه خوارزميه ذكاء"
]
vec_v1 = TfidfVectorizer()
mat_v1 = vec_v1.fit_transform(corpus_v1)

print("4. Représentation TF-IDF (Document 1) :")
df_v1 = pd.DataFrame(mat_v1.toarray(), columns=vec_v1.get_feature_names_out())
print(df_v1.iloc[0].sort_values(ascending=False).to_string())
```

## Version 2 : Pipeline NLP Arabe — Analyse Détaillée

### 1. Tokenisation (التقطيع)
Le texte brut arabe est segmenté en unités minimales appelées **tokens**, après normalisation orthographique.

* **Observation :** Les diacritiques (الشكل) sont supprimés — "يَدْرُسُ" devient "يدرس". Les différentes formes de l'alif (إ، أ، آ) sont unifiées en (ا).
* **Objectif :** Éviter que la machine traite "الحاسوب" et "الحاسوبُ" (avec diacritique) comme deux mots différents. La normalisation garantit une représentation cohérente.

### 2. Retrait des Stopwords (حذف الكلمات الوظيفية)
On élimine les mots grammaticaux arabes qui n'apportent pas de sens discriminant.

* **Observation :** Les mots comme `['في', 'كيفية', 'و']` disparaissent du texte.
* **Objectif :** Réduire le "bruit lexical". En arabe, les prépositions (في، من، على، إلى) et les conjonctions (و، أو، ثم) sont extrêmement fréquentes mais ne définissent pas le sujet du texte. On ne conserve que les **mots pleins** (أسماء، أفعال دالة).

### 3. Racinisation — Stemming (استخراج الجذر)
C'est l'étape de normalisation morphologique. L'**ISRIStemmer** (Information Science Research Institute) ramène chaque mot à sa **racine trilitère** arabe.

* **Transformations typiques :**
   * `معالجة` (traitement, nom verbal) → `عالج` (racine verbale)
   * `الطبيعية` (adjectif féminin défini) → `طبيع` (racine)
   * `علوم` (pluriel) → `علم` (singulier / racine)
* **Objectif :** Regrouper toutes les formes morphologiques d'un mot sous une même représentation. Si un autre texte utilise "طبيعي", "الطبيعي", "طبيعيًا", le modèle reconnaîtra le même concept.

> **Note sur l'arabe vs le français :** En français, on utilise la lemmatisation (forme dictionnaire). En arabe, le stemming est plus courant car la morphologie est plus complexe — un même racine peut générer des dizaines de formes.

### 4. TF-IDF — Pondération Mathématique (التمثيل الرياضي)
On transforme les mots en vecteurs numériques. Le score TF-IDF mesure l'**importance relative** d'un mot dans votre document par rapport au corpus entier.

**Pourquoi certains scores sont-ils plus élevés ?**

Le score dépend de deux facteurs :
1. **TF (Term Frequency)** : Combien de fois le mot apparaît dans votre document.
2. **IDF (Inverse Document Frequency)** : À quel point le mot est **rare** dans les autres documents du corpus.

* **Score élevé** pour "جامع" (université) ou "معالج" (traitement) : Ces mots sont **spécifiques** à votre document et n'apparaissent probablement pas dans les autres textes de comparaison → ils caractérisent fortement votre document.
* **Score moyen** pour "لغه" ou "حاسوب" : Ces mots sont importants, mais s'ils apparaissent aussi dans les autres documents du corpus, leur score IDF baisse.
* **Score 0.00** pour certains mots : Ces racines ne sont tout simplement pas présentes dans votre phrase de test, même si elles font partie du vocabulaire global du corpus.

**En bref :** Votre phrase arabe n'est plus du texte pour la machine — c'est un **point dans un espace vectoriel multidimensionnel**. Plus le score est élevé, plus le mot est considéré comme un mot-clé représentatif de votre document.

```python
import re
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem.isri import ISRIStemmer
from sklearn.feature_extraction.text import TfidfVectorizer

# Outils arabes
arabic_stopwords = set(stopwords.words('arabic'))
stemmer_ar = ISRIStemmer()

# ===========================================================
# STEP 1 : Normalisation & Tokenisation
# ===========================================================
def normalize_arabic(text):
    """Normalise le texte arabe avant tokenisation."""
    # Supprimer les diacritiques (الحركات)
    text = re.sub(r'[\u0610-\u061A\u064B-\u065F]', '', text)
    # Unifier les formes de l'alif
    text = re.sub(r'[إأآا]', 'ا', text)
    # Unifier ya' final
    text = re.sub(r'ى', 'ي', text)
    # Normaliser ta' marbuta
    text = re.sub(r'ة', 'ه', text)
    # Garder uniquement les caractères arabes et espaces
    text = re.sub(r'[^\u0600-\u06FF\s]', '', text)
    return text.strip()

texte_norm_v2 = normalize_arabic(phrase_test_ar)
tokens_bruts_v2 = texte_norm_v2.split()

print("=" * 55)
print(f"TEXTE ORIGINAL : {phrase_test_ar}")
print(f"TEXTE NORMALISÉ : {texte_norm_v2}")
print("=" * 55)
print(f"\n1. Tokenisation (après normalisation) :")
for i, t in enumerate(tokens_bruts_v2):
    print(f"   [{i}] '{t}'")

# ===========================================================
# STEP 2 : Retrait des Stopwords
# ===========================================================
tokens_filtres_v2 = [t for t in tokens_bruts_v2 if t not in arabic_stopwords]
stopwords_supprimes_v2 = [t for t in tokens_bruts_v2 if t in arabic_stopwords]

print(f"\n2. Après retrait des Stopwords :")
print(f"   Tokens conservés : {tokens_filtres_v2}")
print(f"   Mots supprimés   : {list(set(stopwords_supprimes_v2))}")
print(f"   Réduction        : {len(tokens_bruts_v2)} → {len(tokens_filtres_v2)} tokens")

# ===========================================================
# STEP 3 : Stemming (Racinisation ISRIStemmer)
# ===========================================================
racines_v2 = [stemmer_ar.stem(t) for t in tokens_filtres_v2]

print(f"\n3. Racinisation (ISRIStemmer) :")
print(f"   {'Mot original':<25} → {'Racine':<20}")
print(f"   {'-'*45}")
for mot, racine in zip(tokens_filtres_v2, racines_v2):
    print(f"   {mot:<25} → {racine:<20}")

# ===========================================================
# STEP 4 : TF-IDF — Vectorisation
# ===========================================================
corpus_v2 = [
    " ".join(racines_v2),          # Notre document principal
    "طالب جامعه حاسوب تطوان",     # Document de comparaison 1
    "لغه طبيعيه خوارزميه ذكاء"    # Document de comparaison 2
]

vec_v2 = TfidfVectorizer()
mat_v2 = vec_v2.fit_transform(corpus_v2)

df_v2 = pd.DataFrame(
    mat_v2.toarray(),
    columns=vec_v2.get_feature_names_out(),
    index=["Notre texte", "Doc comparaison 1", "Doc comparaison 2"]
)

print(f"\n4. Représentation TF-IDF — Notre document (scores décroissants) :")
scores = df_v2.iloc[0].sort_values(ascending=False)
for mot, score in scores.items():
    barre = '█' * int(score * 20)
    print(f"   {mot:<20} {score:.4f}  {barre}")

#print("\n--- Tableau complet (tous documents) ---")
#print(df_v2.to_string())
```

```python
import re
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem.isri import ISRIStemmer
from sklearn.feature_extraction.text import TfidfVectorizer

# Chargement des outils arabes
arabic_stopwords = set(stopwords.words('arabic'))
stemmer_ar = ISRIStemmer()

def normalize_arabic(text):
    """Normalise le texte arabe : supprime diacritiques, unifie les variantes."""
    text = re.sub(r'[\u0610-\u061A\u064B-\u065F]', '', text)  # Diacritiques
    text = re.sub(r'[إأآا]', 'ا', text)                        # Alifs
    text = re.sub(r'ى', 'ي', text)                              # Ya' final
    text = re.sub(r'ة', 'ه', text)                              # Ta' marbuta
    text = re.sub(r'[^\u0600-\u06FF\s]', '', text)             # Non-arabe
    return text.strip()

def trace_nlp_pipeline_ar(text):
    """
    Applique et trace un pipeline NLP complet sur un texte arabe :
    1. Normalisation & Tokenisation
    2. Retrait des Stopwords
    3. Racinisation (ISRIStemmer)
    Retourne le texte raciné, prêt pour TF-IDF.
    """
    print(f"--- TEXTE BRUT ---\n'{text}'\n")

    # 1. NORMALISATION & TOKENISATION
    texte_norm = normalize_arabic(text)
    tokens_bruts = texte_norm.split()
    print(f"1. Tokenisation (normalisé) :\n   {tokens_bruts}\n")

    # 2. FILTRAGE — Stopwords arabes
    tokens_filtres = [t for t in tokens_bruts if t not in arabic_stopwords]
    supprimes = [t for t in tokens_bruts if t in arabic_stopwords]
    print(f"2. Après retrait des Stopwords :")
    print(f"   {tokens_filtres}")
    print(f"   (Mots supprimés : {list(set(supprimes))})\n")

    # 3. RACINISATION — ISRIStemmer
    # Ramène chaque mot à sa racine trilitère arabe
    racines = [stemmer_ar.stem(t) for t in tokens_filtres]
    print(f"3. Racinisation (Racines trilitères) :")
    print(f"   {racines}\n")

    return " ".join(racines)

# --- EXEMPLE DÉTAILLÉ avec la phrase extraite du PDF ---
phrase_test_ar = "يدرس طلاب علوم الحاسوب في جامعة عبد الملك السعدي كيفية بناء أنظمة معالجة اللغات الطبيعية"
texte_propre_ar = trace_nlp_pipeline_ar(phrase_test_ar)

# --- 4. ZOOM SUR TF-IDF (Vectorisation) ---
# Mini-corpus pour que le calcul TF-IDF ait un sens comparatif
corpus = [
    texte_propre_ar,
    "طالب جامعه حاسوب تطوان",
    "لغه طبيعيه خوارزميه ذكاء"
]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(corpus)

# Affichage des scores pour le premier document
print("4. Représentation Numérique (TF-IDF) pour le document 1 :")
df_tfidf = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out()
)
print(df_tfidf.iloc[0].sort_values(ascending=False))
```
